## Matplotlib example Workflow

creating: 
- line lpot
- scatter plot

- Histogram

```python
# 0. import matplotlib and get it ready for plotting in Jupyter
%matplotlib inline
import matplotlib.pyplot as plt

# 1. prepare data
x = [1,2,3,4]
y = [11,22,33,44]

# 2. Setup plot 
fig, ax = plt.subplots(figsize = (10,10)) #(width,height)

# 3. Plot data
ax.plot(x,y)

# 4. Customize plot
ax.set(title = "Simple Plot",
       xlabel = "x-axis",
       ylabel = "y-axis")

# 5. Save and show (save the whole figure)
fig.savefig("sample-plot.png")
```

```python
# making a plot from dictionary 

nut_butter_prices = {"Almond butter": 10,
                     "Peanut butter": 8,
                     "Cashew butter": 12}
fig, ax = plt.subplots()
ax.bar(nut_butter_prices.keys(), nut_butter_prices.values())
ax.set(title="Klarence's Nut Butter Store",
      ylabel="Price ($)");
```

```python
fig, ax = plt.subplots()
#barh: horizontal bar
ax.barh(list(nut_butter_prices.keys()),list(nut_butter_prices.values()))
ax.set(title="Klarence's Nut Butter Horizontal Store",
      ylabel="Price (Pesos)")
```

### creating multiple plots at the same time 


### Two Options for subplots 

```python
#Subplot Option 1: 

fix, ((ax1, ax2), (ax3,ax4)) = plt.subplots(nrows = 2,
                                          ncols=2,
                                          figsize =(10,5))

#plotting to each different axis

ax1.plot(x,x/2);
ax2.plot(x, x**2);
ax3.scatter(np.random.randn(1000), np.sin(x));
ax4.bar(nut_butter_prices.keys(), nut_butter_prices.values())
```

```python
#Subplot Option 2:

fig, ax = plt.subplots(nrows=2,
                       ncols= 2,
                       figsize = (10,5))

#plot to each different index

ax[0,0].plot(x,x**2)
ax[0,1].plot(x,x//2)
ax[1,0].plot(x, np.sin(x))
ax[1,1].scatter(np.random.random(10),np.random.random(10))
```

## which to use? (pyplot vs matplotlib OO method)

* When plotting something, pylot is okay 
* If more Advanced, use OO method

```python
#filtering the data

over_50 = heart_disease[heart_disease["age"]>50]
over_50.head()
```

### Pyplot Method

```python
#graph of age and cholesterol but color it with Target Col

over_50.plot(kind='scatter',
            x='age',
            y='chol',
            c='target'); #color
```

## OO Method

```python
#OO method
fig, ax = plt.subplots(figsize = (10,6))
over_50.plot(kind='scatter',
            x='age',
            y='chol',
            c='target',
            ax=ax);

#ax.set_xlim([45,100]); 
#this sets the limits. In this case, the x axis will start on 45 and stop showing at 100
```

```python
# OO method from scratch
fig, ax = plt.subplots(figsize= (10,6))

# Plotting the Data
scatter = ax.scatter(x=over_50["age"],
                     y=over_50["chol"],
                     c=over_50["target"]
                    )

#customizing the plot 
ax.set(title = "Heart Disease and Cholesterol Levels",
       xlabel = "Age",
       ylabel = "Cholesterol")

#adding a Legend
ax.legend(*scatter.legend_elements(), title="Target");

#adding a horizontal line
ax.axhline(over_50["chol"].mean(),
          linestyle = "dashed"); 
```

## 

```python
#creating a subplot for heart disease and thalach 
fig, ax = plt.subplots(figsize=(10,6))

# Plotting the Data
scatter = ax.scatter(x = over_50["age"],
                    y = over_50["thalach"],
                    c = over_50["target"]);
#Customizing the plot 

ax.set(title= "Heart Disease and Thalac",
      xlabel="Age",
      ylabel= "Thalac")
#adding a legend 
ax.legend(*scatter.legend_elements(), title="Target");


#adding a mean line
ax.axhline(over_50['thalach'].mean(),
          linestyle = "dashed"); 
```

```python
#Subplot of chol, age, thalac 

fig, (ax0, ax1) = plt.subplots(nrows = 2,
                              ncols = 1,
                              figsize=(10,10),
                              sharex = True)
#note You may use sharex to share the xlines with the other plot


#adding data to ax0 
scatter = ax0.scatter(x = over_50["age"], 
                      y = over_50["chol"],
                      c = over_50["target"]) 

#customizing ax0 
ax0.set(title="Heart Disease and Cholesterol Levels", 
        xlabel = "age",
        ylabel = "chol")

#adding Legend to ax0 
ax0.legend(*scatter.legend_elements(), title= "target")

#adding a dashed line
ax0.axhline(over_50['chol'].mean(), 
          linestyle ="dashed")

#adding data to ax1
scatter1 = ax1.scatter(x = over_50["age"],
                       y= over_50["thalach"],
                       c = over_50["target"])
#Customizing the ax1
ax1.set(title= "Heart Disease and Thalac", 
       xlabel = "Age",
       ylabel = "Thalac")
#adding Legend to ax1 

ax1.legend(*scatter1.legend_elements(), title = "target")

#adding a mean line 

ax1.axhline(y = over_50["thalach"].mean(),
            linestyle = "dashed");


#adding the title to the figure 
fig.suptitle("Heart Disease Analysis", fontsize = 16, fontweight="bold")
```

## Changing the style within the style
### Customizing the y and x axis limitations

```python
#Subplot of chol, age, thalac 
fig, (ax0, ax1) = plt.subplots(nrows = 2,
                              ncols = 1,
                              figsize=(10,10),
                              sharex = True)
#note You may use sharex to share the xlines with the other plot


#adding data to ax0 
scatter = ax0.scatter(x = over_50["age"], 
                      y = over_50["chol"],
                      c = over_50["target"],
                     cmap="winter") #changes the color scheme 

#customizing ax0 
ax0.set(title="Heart Disease and Cholesterol Levels", 
        ylabel = "chol",
        xlim= (50,80),
       ylim= (100,600))

#adding Legend to ax0 
ax0.legend(*scatter.legend_elements(), title= "target")

#adding a dashed line
ax0.axhline(over_50['chol'].mean(), 
          linestyle ="dashed",
           color= "red")

#adding data to ax1
scatter1 = ax1.scatter(x = over_50["age"],
                       y= over_50["thalach"],
                       c = over_50["target"],
                       cmap="winter") #changes the color scheme 

#Customizing the ax1
ax1.set(title= "Heart Disease and Thalac", 
       xlabel = "Age",
       ylabel = "Thalac",
       xlim= (50,80),
       ylim= (60,200))
#adding Legend to ax1 

ax1.legend(*scatter1.legend_elements(), title = "target")

#adding a mean line 

ax1.axhline(y = over_50["thalach"].mean(),
            linestyle = "dashed",
           color = "red");


#adding the title to the figure 
fig.suptitle("Heart Disease Analysis", fontsize = 16, fontweight="bold")
```
